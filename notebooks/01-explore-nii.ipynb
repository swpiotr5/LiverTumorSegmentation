```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from matplotlib import colors
import gc
import time

segm_dir = '../data/segmentations'
volumes_dir = '../data/volumes'
destination = '../data/prepared_data'

volume_files = sorted(os.listdir(volumes_dir))
mask_files = sorted(os.listdir(segm_dir))

# Mapa kolorów: 0 - tło (czarny), 1 - wątroba (czerwony), 2 - guz (żółty)
cmap = colors.ListedColormap(['black', 'red', 'yellow'])
bounds = [0, 0.5, 1.5, 2.5]
norm = colors.BoundaryNorm(bounds, cmap.N)

for volume_file, mask_file in zip(volume_files, mask_files):
    volume_path = os.path.join(volumes_dir, volume_file)
    mask_path = os.path.join(segm_dir, mask_file)
    #tworzymy podkatalogi
    patient_id = volume_file.split('-')[1].replace('.nii', '')
    base_dir = f'../data/prepared/person_{patient_id}'
    
    if os.path.exists(base_dir):
        print(f"{base_dir} już istnieje – pomijam")
        continue
    else:
        os.makedirs(base_dir, exist_ok=True)
        
    for sub in ['ct', 'masks', 'liver', 'tumor', 'liver_tumor', 'background_only']:
        os.makedirs(os.path.join(base_dir, sub), exist_ok=True)
    ct_dir = os.path.join(base_dir, 'ct')
    mask_dir = os.path.join(base_dir, 'masks')
    background_dir = os.path.join(base_dir, 'background_only')
    liver_dir = os.path.join(base_dir, 'liver')
    tumor_dir = os.path.join(base_dir, 'tumor')
    liver_tumor_dir = os.path.join(base_dir, 'liver_tumor')

    for sub in ['ct', 'mask']:
        os.makedirs(os.path.join(liver_dir, sub), exist_ok=True)
        os.makedirs(os.path.join(tumor_dir, sub), exist_ok=True)
        os.makedirs(os.path.join(liver_tumor_dir, sub), exist_ok=True)
        os.makedirs(os.path.join(background_dir, sub), exist_ok=True)
    
    mask = nib.load(mask_path).get_fdata()
    volume = nib.load(volume_path).get_fdata()

    print(patient_id, base_dir, ct_dir, mask_dir)
    for i in range(mask.shape[2]):
        ct_path = f'{ct_dir}/slice_{i:03d}.npy'
        mask_path = f'{mask_dir}/slice_{i:03d}.npy'
        
        
        slice_mask = mask[:, :, i]
        ct_slice = volume[:, :, i]

        np.save(ct_path, ct_slice)
        np.save(mask_path, slice_mask)

        unique_vals = np.unique(slice_mask) 
        print(f"Slice {i}: unique values = {unique_vals}")
        
        
        if 1 in unique_vals and 2 not in unique_vals:
            print(f"-> Slice {i} zawiera tylko tło i wątrobę")
            ct_path = f'{liver_dir}/ct/slice_{i:03d}.npy'
            mask_path = f'{liver_dir}/mask/slice_{i:03d}.npy'
            np.save(ct_path, ct_slice)
            np.save(mask_path, slice_mask)
            
        elif 2 in unique_vals and 1 not in unique_vals:
            print(f"-> Slice {i} zawiera tylko guz nowotworowy")
            ct_path = f'{tumor_dir}/ct/slice_{i:03d}.npy'
            mask_path = f'{tumor_dir}/mask/slice_{i:03d}.npy'
            np.save(ct_path, ct_slice)
            np.save(mask_path, slice_mask)
            
        elif 2 in unique_vals and 1 in unique_vals:
            print(f"-> Slice {i} zawiera wątrobę i guz nowotworowy")
            ct_path = f'{liver_tumor_dir}/ct/slice_{i:03d}.npy'
            mask_path = f'{liver_tumor_dir}/mask/slice_{i:03d}.npy'
            np.save(ct_path, ct_slice)
            np.save(mask_path, slice_mask)
        else:
            print(f"-> Slice {i} zawiera tylko tło.")
            ct_path = f'{background_dir}/ct/slice_{i:03d}.npy'
            mask_path = f'{background_dir}/mask/slice_{i:03d}.npy'
            np.save(ct_path, ct_slice)
            np.save(mask_path, slice_mask)

        del ct_slice, slice_mask
        gc.collect()
```
